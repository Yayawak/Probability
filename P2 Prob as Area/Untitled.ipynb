```python
import random 
n = 100 # จำนวนลูกเต๋าที่จะโยน
dices_face = [] # like {1,5,3,2,6,...}
for i in range(n):
    dices_face += [random.randint(1,6)]
print(dices_face)
```

```python
from collections import Counter
print(Counter(dices_face))
```

```python
p = [1,2,2,2,2,2,2,2,3,4,5,5,6] # weigh at 2 the most
n = 100
d_faces = [] # d = dice
for i in range(n):
    d_faces.append(random.choice(p))
print(d_faces)
print(Counter(d_faces))


```

```python
import matplotlib.pyplot as plt

plt.hist(d_faces, )
```

```python
import numpy as np

plt.figure(figsize=(6,6),)
ax = plt.axes()
# plt.axvspan(0,0.5,alpha=.9, fc='r')
# plt.axvspan(.5, 1,alpha=.9, fc='b')

l = plt.Rectangle([0,0],.5,1, fc='b', ec='b', )
r = plt.Rectangle([.5,0],.5,1, fc='r', ec='b', )

plt.grid(ls='--')

plt.text(l.get_x(),l.get_y(), 'red')
ax.add_patch(l)
ax.add_patch(r)
plt.show()
```
